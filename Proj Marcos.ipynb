```python
import pandas as pd
import matplotlib.pyplot as plt

#Abrindo xlsx e lendo
xlsx = pd.ExcelFile('/home/lu-ks/Downloads/Python3/dados_IGM-CFA_2016.xlsx')
```

```python
#Verificando quantas planilhas há no arquivo e seus nomes

xlsx.sheet_names
```

```python
#Observando 10 primeiras linhas para se preparar a limpar os dados

xlsx.parse(sheet_name='Ranking', skiprows=0).head(4)
```

```python
#Criando dataframe onde a primeira linha foi pulada para criação do cabeçalho

df1 = xlsx.parse(sheet_name='Ranking', skiprows=1)
df1
```

```python
#Testando se está tudo bem com os dados e observando variáveis uteis
df1.info()
```

```python
#Selecionando variaveis uteis
df = pd.DataFrame(df1, columns=['Índice Firjan de Gestão Fiscal (2015)','IDHM','IGM - 2016','Mortalidade infantil','Gastos saúde e educação per capita','Qualidade habitacional','Educação','População 2016 (IBGE)','PIB per capita (R$) 2013 (IBGE)'])
df.head(4)
```

```python
#Avaliando correlação entre variáveis
df.corr()
```

```python
print(df.plot.scatter('IDHM','Mortalidade infantil'))
```

```python
print(df.plot.scatter('IDHM','Qualidade habitacional'))
```

```python
print(df.plot.scatter('IDHM','Educação'))
```

```python
print(df.plot.scatter('IDHM','PIB per capita (R$) 2013 (IBGE)'))
```

```python
print(df.plot.scatter('Qualidade habitacional','PIB per capita (R$) 2013 (IBGE)'))
```

```python
print(df.plot.scatter('Gastos saúde e educação per capita','PIB per capita (R$) 2013 (IBGE)'))
```

```python
print(df.plot.scatter('IGM - 2016','PIB per capita (R$) 2013 (IBGE)'))
```
